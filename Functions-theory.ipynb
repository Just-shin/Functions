2. Explain the concept of function arguments and parameters in Python.


Answer - 
    
    In Python, the terms arguments and parameters are often used interchangeably, but they have distinct meanings:
    Parameters >> 
        Parameters are the variables listed inside the parentheses in the function definition. They act as placeholders for the values that the function will operate on. Parameters allow functions to be generalized and reusable. They enable us to write code that can work with different values each time the function is called. example :
        def greet(name):   ##here name is parameter.
            print(f"Hello, {name}!")
        Arguments >> 
        Arguments are the actual values passed to the function when it is called. These values are assigned to the corresponding parameters. It might be a variable, value or object passed to a function or method as input. Arguments provide the specific data that the function needs to perform its operations. Example :
        greet("Alice") #here "Alice" is an argument passed to the 'greet' function.
        Types of Arguments in Python:
        Positional Arguments:
        These are passed to the function in the order they are defined in the parameter list.
        Keyword Arguments:
        These are passed with the parameter name, allowing you to specify the value for a particular parameter explicitly. The order does not matter.
        Default Arguments:
        These are parameters with predefined values that are used if no corresponding argument is provided during the function call.
        Variable-Length Arguments:
        These allow a function to accept an arbitrary number of arguments, using *args for positional arguments and **kwargs for keyword arguments.
    Differences between them >> 
    Parameters exist only within the function definition and act as placeholders, while arguments are the values passed from outside the function when it's called. Parameters define the type and number of inputs a function can accept, while arguments provide the actual data for those inputs.


```python
def calculate_total(price, quantity=1, discount=0):
    total = price * quantity * (1 - discount)
    return total

# Using positional arguments
result1 = calculate_total(10, 2)

# Using keyword arguments
result2 = calculate_total(price=10, discount=0.1)

# Using default arguments
result3 = calculate_total(10)
# Using variable-length arguments
def add(*args):
    return sum(args)
result = add(1, 2, 3, 4)  # *args collects all positional arguments into a tuple
```

3. What are the different ways to define and call a function in Python?

Answer - 
   
   
    In Python, there are several ways to define and call functions, each serving different purposes and offering various levels of flexibility. Here are some common methods:

```python
'''DEFINING FUNCTIONS'''
#Standard Function Definition >> def keyword followed by the function name and parentheses.
def greet(name):
    print(f"Hello, {name}!")

#Lambda Functions >>  to create small anonymous functions.
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8

#Nested Functions >> Define a function inside another function.
def outer_function(text):
    def inner_function():
        print(text)
    inner_function()

outer_function("Hello from the inner function!")

#Functions with Default Arguments >> Define default values for parameters.
def greet(name="Uncle!"):
    print(f"Hello, {name}!")

greet()        # Output: Hello, Uncle!
greet("Niece!") # Output: Hello, Niece!


'''CALLING FUNCTIONS'''
#Direct Call >> Call the function by its name followed by parentheses.
def greet(name):
    print(f"Hello, {name}!")

greet("Shin")

#Using Positional Arguments >>  Pass arguments in the same order as the parameters.
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8

#Using Keyword Arguments >> Pass arguments by explicitly naming each parameter.
def greet(first_name, last_name):
    print(f"Hello, {first_name} {last_name}!")

greet(last_name="Inu", first_name="Shiba")

#Calling Functions with Variable-length Arguments >> 
def add(*args):
    return sum(args)

result = add(1, 2, 3, 4)
print(result)  # Output: 10

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Cat", age=30)

#Calling Functions from Another Function >> Call one function from within another function.
def greet(name):
    print(f"Hello, {name}!")

def welcome():
    greet("Shin")

welcome()  # Output: Hello, Shin!
```

4. What is the purpose of the 'return' statement in a Python function?

Answer - 

    The Python 'return' statement marks the end of a function and specifies the value or values to pass back from the function. Return statements can return data of any type, including integers, floats, strings, lists, dictionaries, and even other functions. 
    In Python, the return statement serves two primary purposes within a function:
    1. Terminating/Exiting the Function: When the return statement is encountered within a function, the function's execution immediately stops, and control returns to the code that called the function.
    2. Sending Values Back: The return statement allows a function to send a value (or multiple values) back to the caller. This value can then be used by the code that called the function.
    3. Returning None : If no value is specified, the function returns None by default.

```python
#Exiting a Function >> The return statement terminates the execution of a function and transfers control back to the calling code.
def greet(name):
    return f"Hello, {name}!"
    print("This will not be printed")

message = greet("Shin")
print(message)  # Output: Hello, Shin!

#Returning a Value >> It allows a function to send a value back to the caller, which can then be used for further computation or processing.
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8

#Returning Multiple Values >> Python functions can return multiple values as a tuple, which can be unpacked by the caller.
def get_coordinates():
    return (10, 20)

x, y = get_coordinates()
print(x, y)  # Output: 10 20

#Returning None >> If no value is specified, the function returns None by default.
def greet(name):
    print(f"Hello, {name}!")

result = greet("Shin")
print(result)  # Output: Hello, Shin! \n None
```

5. What are iterators in Python and how do they differ from iterables?

Answer - 

    In Python, an "iterable" is any object that can be looped over using a "for" loop, like lists, tuples, or strings, while an "iterator" is an object that provides a mechanism to access elements of an iterable one at a time, essentially acting as a pointer that moves through the iterable, returning each element sequentially when you call its __next__() method. All iterables can be used to create iterators, but not all iterators are iterables.

```python
#Iterables >> Common iterables include lists, tuples, strings, dictionaries, and sets.
my_list = [1, 2, 3]
for item in my_list:
    print(item)

#Iterators >> must implement two methods: __iter__() and __next__().
my_list = [1, 2, 3]
iterator = iter(my_list)
print(next(iterator))
print(next(iterator))
print(next(iterator))

"""Key Differences >> 
Creation >>
Iterable: Any object with an __iter__() method.
Iterator: Any object with both __iter__() and __next__() methods.
Usage >> 
Iterable: Used to create an iterator. Can be looped over using a for loop.
Iterator: Used to fetch items one at a time using the next() function.
State >> 
Iterable: Does not maintain any state.
Iterator: Maintains state, keeping track of where it is during iteration.
Exhaustion >>
Iterable: Can be re-iterated multiple times.
Iterator: Can be exhausted; once all items are fetched, it cannot be reused without reinitializing."""
```

6. Explain the concept of generators in Python and how they are defined?

Answer - 

    Generators in Python are a special type of iterable that allow us to iterate over a sequence of values providing a faster and easier way to create iterators without storing the entire sequence in memory. They are particularly useful for working with large datasets or streams of data where it is impractical to load everything into memory at once. It's like a normal function, but whenever it needs to generate a value, it does so with the yield keyword rather than return.


```python
#Generators functions >> Defined like a normal function using the def keyword, but instead of returning a value,they use the yield keyword to produce a series of values.
#Each time 'yield' is called, the functionâ€™s state is saved, and it can be resumed later from where it left off.
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

counter = count_up_to(6)
for number in counter:
    print(number)

#Generator Expressions >> A concise way to create generators using a syntax similar to list comprehensions but with parentheses instead of square brackets
squares = (x * x for x in range(10))
for square in squares:
    print(square)
```

7. What are the advantages of using generators over regular functions ?

Answer - 
    Generators offer several advantages over regular functions, particularly when dealing with large datasets or streams of data, additionally, generators can improve code readability by clearly defining iterative processes and can be used to create complex data streams in a more manageable way.
    
    Advantages of Generators  >>
    ##Memory Efficiency: Generators do not store the entire sequence in memory; they generate each value on-the-fly.
    ##Lazy Evaluation: Values are produced only when requested, which can lead to performance improvements.
    ##Simplified Code: Generators can simplify code that needs to produce a sequence of values, especially when dealing with complex logic or large datasets.

```python
#Key advantages of generators >> 

#Memory efficiency(Lazy evaluation) >> Generators produce items one at a time and only when required, which means they do not store the entire sequence in memory.
#Beneficial for large data sets or when dealing with potentially long computations.
def generate_numbers():
    for i in range(1000000):
        yield i

gen = generate_numbers()
print(next(gen))
print(next(gen))

#Improved Performance >> 
#Reduced Overhead: Since generators yield items one at a time, they can start producing results immediately without waiting to generate the entire sequence. This can lead to faster execution times.
def generate_squares(n):
    for i in range(n):
        yield i * i

squares = generate_squares(10)
for square in squares:
    print(square)

#Simplified Code >> 
#Cleaner Syntax: Generators can simplify complex iteration logic by using the yield keyword, making the code more readable and maintainable.
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
for _ in range(10):
    print(next(fib)) #two geerators first simple and second using for loop.

#Infinite Sequences >> Generators can easily represent infinite sequences by pausing and resuming execution using the "yield" keyword. 
#Handling Infinite Data: Generators can produce an infinite sequence of values without running out of memory, which is not feasible with regular functions that return lists.
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1

gen = infinite_sequence()
print(next(gen))
print(next(gen))

#State Retention >> Maintaining State: Generators maintain their state between successive calls, which can be useful for complex stateful computations.
def countdown(n):
    while n > 0:
        yield n
        n -= 1

cd = countdown(5)
for number in cd:
    print(number)

#Generators provide a powerful and efficient way to handle large datasets, complex iteration logic, and infinite sequences, all while keeping memory usage low and code clean.
```

    A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions created with the def keyword, lambda functions are typically used for short, simple operations and are defined in a single line line function for a short-term operation, particularly when passing a function as an argument to other functions like map, filter, or sorted where a full function definition might be unnecessary
    Basic syntax - lambda arguments: expression.
    #Arguments: The parameters passed to the function.
    #Expression: A single expression that is evaluated and returned.
    
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8


    

```python
#TYPICAL USES > Lambda functions are often used in situations where a small, throwaway function is needed for a short period.

#Inline Functions >> Lambda functions can be used as arguments to higher-order functions like map(), filter(), and reduce().
numbers = [1, 2, 3, 4, 5]
squares = map(lambda x: x * x, numbers)
print(list(squares))

#Condition Checking Using Python lambda function >>
format_numeric = lambda num: f"{num:e}" if isinstance(num, int) else f"{num:,.2f}"

print("Int formatting:", format_numeric(1000000))
print("float formatting:", format_numeric(999999.789541235))

#List comprehension >> 
is_even_list = [lambda arg=x: arg * 10 for x in range(1, 5)]
for item in is_even_list:
    print(item())

#Lambda Function with if-else >>
Max = lambda a, b : a if(a > b) else b
print(Max(1, 2))

#lambda() Function with filter() >> 
li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]

final_list = list(filter(lambda x: (x % 2 != 0), li))
print(final_list)

#Sorting and Ordering >> can be used as the key argument in sorting functions.
points = [(1, 2), (3, 1), (5, -1)]
points_sorted = sorted(points, key=lambda point: point[1])
print(points_sorted)

#Event Handling >> In GUI programming, lambda functions can be used to handle events in a concise manner.
'''import tkinter as tk
root = tk.Tk()
button = tk.Button(root, text="Click me", command=lambda: print("Button clicked!"))
button.pack()
root.mainloop()'''

#Functional Programming >> Lambda functions are often used in functional programming paradigms where functions are passed as arguments or returned as values.
def make_incrementor(n):
    return lambda x: x + n

increment_by_2 = make_incrementor(2)
print(increment_by_2(5))

#Advantages >> 
#Conciseness: Lambda functions provide a concise way to define small functions.
#Readability: They can make code more readable by reducing the need for boilerplate code.
#Limitations >>
#Single Expression: Lambda functions are limited to a single expression and cannot contain multiple statements or complex logic.
#Readability: Overuse of lambda functions can sometimes make code harder to read and understand.
```

9. Explain the purpose and usage of the 'map()' function in Python. 

Answer - 
    
    The map() function in Python is a built-in function that allows to apply a given function to each item of an iterable (like a list, tuple, etc.) and return a map object (which is an iterator) containing the results. This is particularly useful for transforming data efficiently and concisely.
    Purpose:
        To transform elements of an iterable without using explicit loops.
        To apply a function to each item in an iterable, producing a new iterable.
    Syntax - map(function, iterable, ...)
    function: The function to apply to each item.
    iterable: One or more iterables whose items will be passed to the function.

#Example - 
def double(n):
    return n * 2

numbers = [1, 2, 3, 4]
result = map(double, numbers)
print(list(result))  # Output: [2, 4, 6, 8]

Advantages >> 
Efficiency: map() can be more efficient than using a for loop, especially for large datasets.
Readability: It can make the code more readable by clearly expressing the transformation being applied to each item.
Functional Programming: Supports a functional programming style, making it easier to apply transformations to data.


```python
#Using Lambda Functions >> Square each number using a lambda function
numbers = [1, 2, 3, 4, 5] 
squared_numbers = map(lambda x: x**2, numbers)
squared_numbers_list = list(squared_numbers) # Convert the iterator to a list
print(squared_numbers_list)

#Multiple Iterables >> Adding corresponding elements of two lists.
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
result = map(lambda x, y: x + y, numbers1, numbers2)
print(list(result))

#Modifying Strings >> Converting a list of strings to a list of lists of characters.
strings = ['sat', 'bat', 'cat']
result = map(list, strings)
print(list(result))

#Conditional Logic >> Doubling even numbers and leaving odd numbers unchanged.
def double_even(num):
    return num * 2 if num % 2 == 0 else num

numbers = [1, 2, 3, 4, 5]
result = map(double_even, numbers)
print(list(result))


### The map() function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.
```

10. What is the difference between map(), reduce(), and filter() functions in Python ?

Answer -
    
    Key Differences >> 
Functionality >> 
        map(): Transforms each element.
        filter(): Selects elements based on a condition.
        reduce(): Aggregates elements to a single value.
Return Type >>
        map() and filter() return iterators.
        reduce() returns a single value.
Use Cases >>
        map(): When you need to apply a function to each element.
        filter(): When you need to filter elements based on a condition.
        reduce(): When you need to combine elements into a single result.

```python
#map() >> 
#Purpose: Applies a given function to each item of an iterable (like a list) and returns a map object (which is an iterator) containing the results.
#Syntax: map(function, iterable, ...)
def double(x):
    return x * 2
numbers = [1, 2, 3, 4]
result = map(double, numbers)
print(list(result)) 
#Use Case: Transforming data by applying a function to each element.

#filter() >>
#Purpose: Filters elements from an iterable based on a function that returns True or False.
#Syntax: filter(function, iterable)
def is_even(x):
    return x % 2 == 0
numbers = [1, 2, 3, 4, 5, 6]
result = filter(is_even, numbers)
print(list(result))
'''Use Case: Selecting elements that meet a certain condition.'''

#reduce() >>
#Purpose: Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value. It is part of the functools module.
#Syntax: reduce(function, iterable[, initializer])
from functools import reduce

def add(x, y):
    return x + y
numbers = [1, 2, 3, 4]
result = reduce(add, numbers)
print(result)
#Use Case: Aggregating elements to produce a single cumulative result.

##In Summary:
#map(): Transforms elements.
#filter(): Selects elements based on a condition.
#reduce(): Combines elements into a single value.
```
